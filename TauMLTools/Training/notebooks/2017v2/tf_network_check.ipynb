```python
import tensorflow as tf
import numpy as np
```

```python
def load_graph(graph_filename):
    with tf.gfile.GFile(graph_filename, 'rb') as f:
        graph_def = tf.GraphDef()
        graph_def.ParseFromString(f.read())

    with tf.Graph().as_default() as graph:
        tf.import_graph_def(graph_def, name="")
    return graph
```

```python
graph = load_graph('../../../../output/networks/2017v2p6/DeepTau2017v2p6_step1_e6.pb')
```

```python
w_tensors = []
for op in graph.get_operations():
    for x in op.values():
        if len(x.shape) > 0 and x.shape[0].value is not None and "FusedBatchNorm" not in x.name \
           and "flatten" not in x.name:
            w_tensors.append(x)
w_tensors
```

```python
with graph.as_default() as gr:
    with tf.Session() as sess:
        weights = sess.run(w_tensors)
```

```python
nan_detected = False
for n in range(len(weights)):
    if np.isnan(weights[n]).any():
        print("{}: NaN detected".format(w_tensors[n].name))
        nan_detected = True
if not nan_detected:
    print("No NaN was detected")
```
